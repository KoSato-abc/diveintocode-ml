**【問題1】べき乗の算術演算子を使用して作成**

```
"""
紙を1回折った時の厚さを計算するコード
"""
THICKNESS = 0.00008
num_folds = 43
folded_thickness = THICKNESS * (2 ** (num_folds )) 
print("厚さ： {}メートル".format(folded_thickness))
```

**【問題2】単位の変換**

```
# メートルをキロメートルに変換して小数点以下2桁で表示する
print("厚さ： {:.2f}万キロメートル".format(folded_thickness/10000000))
```

**【問題3】for文を使用して作成**

```
exponent = 2

for i in range(num_folds-1):
  exponent = exponent * 2
folded_thickness_2 = exponent * THICKNESS

print("厚さ： {:.2f}万キロメートル".format(folded_thickness_2/10000000))
```

**【問題4】計算時間の比較**

*べき乗の場合*

```
import time

start = time.time()
#####
folded_thickness = THICKNESS * (2 ** (num_folds )) 
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

*for文の場合*

```
exponent = 2

start = time.time()
#####
for i in range(num_folds-1):
  exponent = exponent * 2
folded_thickness_2 = exponent * THICKNESS
#####
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

上記の計測結果から
実行速度はべき乗の場合の方が早いことが分かる

**【問題5】リストへの保存**

```

exponent = 2
folded_thickness_list = [THICKNESS]

for i in range(num_folds-1):
  exponent = exponent * 2
  folded_thickness_list.append(THICKNESS * exponent)

len(folded_thickness_list)
```

**【問題6】折れ線グラフの表示**

```
"""
グラフを表示する。タイトルと軸ラベル名付き。
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list)
plt.show()
```

紙の厚さが指数関数的に増加していることが分かる

**【問題7】グラフのカスタマイズ**

*カスタマイズ１*

```
"""
赤い折れ線グラフを表示する。
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list, color='red')
plt.show()
```

*カスタマイズ２*

```
"""
緑の折れ線グラフ（破線）を表示する。
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list, 'r--',color='green')
plt.show()
```

*カスタマイズ３*

```
"""
グレーの折れ線グラフ（破線）を表示する。
"""
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(folded_thickness_list, 'o--', color='grey', alpha=0.3)
plt.show()
```

**【問題】栗まんじゅうが太陽系を覆う日**

```
import math

def calc_full_volume_min( base_volume, full_volume):
  
  i = 0
  now_volume = base_volume

  while True:
    i+=5
    now_volume = now_volume * 2

    if full_volume <= now_volume:
      return i

# print(calc_full_volume_min(4,32))
```

```
# 太陽の高さ　696340km
SUN_HIGHT = 696340

# 太陽系の半径　25000光年
SOLAR_SYS_RAD = 25000 * 9.461e+12

# 太陽系の体積（円柱として計算）
SOLAR_SYS_VOL = SOLAR_SYS_RAD**2 * math.pi * SUN_HIGHT

# 栗饅頭の体積（10**-13Kｍ3）
KURI_MANJU_VOL = 10**-13

# 問題文の計算
print("栗まんじゅうが太陽系を覆うまでにかかる時間は{}分".format(calc_full_volume_min(KURI_MANJU_VOL,SOLAR_SYS_VOL)))
```

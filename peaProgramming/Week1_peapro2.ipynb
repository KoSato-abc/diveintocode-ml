```
THICKNESS = 0.00008
MT_FUJI_HIGH = 3776
min_folds = 0
i = 0

while True:
  i+=1
  folded_thickness = THICKNESS * (2**i)

  if MT_FUJI_HIGH < folded_thickness:
    break

print("富士山の高さを超えるのは{}回折ったとき({}m)".format(i,folded_thickness))
```

**【問題2】任意の厚さに対応した関数を作成**

```
def calc_folded(high, thickness = 0.00008):

  min_folds = 0
  i = 0

  while True:
    i+=1
    folded_thickness = thickness * (2**i)

    if high < folded_thickness:
      return i
```

**【問題3】必要な紙の長さ**

```
import math

# n 折る回数
# t 紙の厚さ
def calc_paper_length(n, t = 0.00008):
  return math.pi*t/6 * (2**n+4) * (2**n-1)
```

```

# 富士山にたどり着くために必要な折る回数と紙の長さ
folded_to_fj = calc_folded(MT_FUJI_HIGH)
length_to_fj = calc_paper_length(folded_to_fj)

# 月までの距離　384400000m
# 月にたどり着くために必要な折る回数と紙の長さ
folded_to_moon = calc_folded(384400000)
length_to_moon = calc_paper_length(folded_to_moon)

# 恒星にたどり着くために必要な折る回数と紙の長さ
folded_to_star = calc_folded(4014190000000000000000)
length_to_star = calc_paper_length(folded_to_star)

print("富士山 折る回数:{}回 紙の長さ:{}m".format(folded_to_fj,length_to_fj))
print("月　　 折る回数:{}回 紙の長さ:{}m".format(folded_to_moon,length_to_moon))
print("恒星　 折る回数:{}回 紙の長さ:{}m".format(folded_to_star,length_to_star))

```

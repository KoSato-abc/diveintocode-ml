```python
"""
col_flg
1 : 試合情報　＋　チーム順位
2 : 試合情報　＋　チーム順位　＋　選手情報
3 : 試合情報　＋　チーム順位　＋　選手情報　＋　スタッツ
4 : 試合情報　＋　チーム順位　＋　スタッツ
pred_flg
1 : 多値分類 ： 勝分敗
2 : 多値分類 ： 得点数（0点、1点、２点、３点以上)
3 : 回帰 ： 得点数
"""
def get_train_test(col_flg, pred_flg, away_flg = False):
    df = df_all.copy()
    col_flg = str(col_flg)
    pred_flg = str(pred_flg)
    # 対象カラムのデータを抽出
    df = get_abs_columns(df, col_flg)
    # train, testに分割（〜2019年をtrain、2020年をtest)
    train, test = my_train_test_split(df, 2020)
    # 
    X_train, y_train = my_x_y_split(train, pred_flg, away_flg)
    X_test, y_test = my_x_y_split(test, pred_flg, away_flg, y_test_flg = True)
    
    return X_train, y_train, X_test, y_test
```

```python
def get_abs_columns(df, col_flg):
    if col_flg=="1":
        columns = columns_1 + columns_2
    elif col_flg=="2":
        columns = columns_1 + columns_2 + columns_3
    elif col_flg=="3":
        columns = columns_1 + columns_2 + columns_3 + columns_4
    else:
        columns = columns_1 + columns_2 + columns_4
    columns = columns + columns_y
    
    return df.loc[:,columns]
```

```python
def my_train_test_split(df, y_year):
    df = df.copy()
    df["year"] = df["年月日"].apply(lambda x : int(str(x)[:4]))
    
    train = df[df["year"] < y_year].drop(columns=["year"])
    test = df[df["year"] == y_year].drop(columns=["year"])
    
    return train, test
```

```python
def my_x_y_split(df, pred_flg, away_flg, y_test_flg = False):
    df = df.copy()
    y_col = ["H_goal", "A_goal", "H_result"]
    x = df.drop(columns=y_col)
    y = df.loc[:,y_col]
    
    y["1"] = y["H_result"]
    if away_flg:
        y["2"] = y["H_goal"].apply(lambda x : 3 if (x > 3 ) else x)
        y["3"] = y["H_goal"]
    else:
        y["2"] = y["A_goal"].apply(lambda x : 3 if (x > 3 ) else x)
        y["3"] = y["A_goal"]
    
    if y_test_flg:
        y = y["1"]
    else:
        y = y[pred_flg]
    return x, y
```

```python
class save_result(list):
    
    def __init__(self):
        self.cols = ["使用カラム", "予測パターン", "正解率"]
        self.df = pd.DataFrame(columns = self.cols)
    
    def save(self, list):
        df = pd.DataFrame([list], columns = self.cols)
        self.df = pd.concat([self.df, df], axis = 0)
        
sr = save_result()
```

・使用データ・・・①試合情報 ＋ チーム順位  
・予測するもの・・・①勝分敗の多値分類  

```python
def do_rightgbm_pred_flg1(col_flg):
    
    pred_flg = 1
    x_train, y_train, x_test, y_test = get_train_test(col_flg, pred_flg)
    x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, train_size=0.75)

    lgb_train = lgb.Dataset(x_train, y_train)
    lgb_eval = lgb.Dataset(x_val, y_val, reference=lgb_train)
    # 学習用パラメータ
    lgbm_params = {
        'objective': 'multiclass',
        'metric': 'multi_logloss',
        'num_class': 3,
    }
    # 学習
    model = lgb.train(lgbm_params, lgb_train, valid_sets=lgb_eval)
    
    y_pred_proba = model.predict(x_test, num_iteration=model.best_iteration)
    y_pred = np.argmax(y_pred_proba, axis=1)
    # Accuracy の計算
    accuracy = sum(y_test == y_pred) / len(y_test)
    print('accuracy:', accuracy)

    sr.save([col_flg, pred_flg, accuracy])
    
#     # feature importanceを表示
#     importance = pd.DataFrame(model.feature_importance()
#                               , index=x_train.columns
#                               , columns=['importance']).sort_values("importance",ascending=False)
#     display(importance)
```

```python
def get_y_pred2(y_pred_proba_home, y_pred_proba_away, y_test):
    
    y_pred = []
    thresholds　= np.arange(0.1, 2, 0.05)
    best_accuracy = 0
    best_threshold = 0
    y_pred_home = np.argmax(y_pred_proba_home, axis=1)
    y_pred_away = np.argmax(y_pred_proba_away, axis=1)
    
    for threshold in thresholds:
        y_pred_tmp = []
        for i in range(len(y_pred_home)):
            if y_pred_home[i]==y_pred_away[i]:
                p_home = y_pred_proba_home[i][y_pred_home[i]]
                p_away = y_pred_proba_away[i][y_pred_away[i]]
                if abs(p_home-p_away) < threshold:
                    y_pred_tmp += [0]
                else:
                    p_home = np.array(y_pred_proba_home[i]).argsort()[-2]
                    p_away = np.array(y_pred_proba_away[i]).argsort()[-2]
                    if p_home > p_away:
                        y_pred_tmp += [1]
                    else:
                        y_pred_tmp += [2]
            elif y_pred_home[i]>y_pred_away[i]:
                y_pred_tmp += [1]
            else:
                y_pred_tmp += [2]
        
        accuracy = sum(y_test == y_pred_tmp) / len(y_test)
        
        if best_accuracy < accuracy:
            best_accuracy = accuracy:
            y_pred = y_pred_tmp
            best_threshold = threshold
    return np.array(y_pred), 
```

```python
def do_rightgbm_pred_flg2(col_flg):
    # 対象のデータを取得
    pred_flg = 2
    x_train, y_train, x_test, y_test = get_train_test(col_flg, pred_flg, False)
    x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, train_size=0.75)
    # モデル作成
    lgb_train = lgb.Dataset(x_train, y_train)
    lgb_eval = lgb.Dataset(x_val, y_val, reference=lgb_train)
    # 学習用パラメータ
    lgbm_params = {
        'objective': 'multiclass',
        'num_class': 4,
        }
    # 学習
    model = lgb.train(lgbm_params, lgb_train, valid_sets=lgb_eval)
    # 推定
    y_pred_proba_home = model.predict(x_test, num_iteration=model.best_iteration)
#     y_pred_home = np.argmax(y_pred_proba, axis=1)
    print(y_pred_proba_home)
    # 対象のデータを取得
    x_train, y_train, x_test, y_test = get_train_test(col_flg, pred_flg, True)
    x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, train_size=0.75)
    # モデル作成
    lgb_train = lgb.Dataset(x_train, y_train)
    lgb_eval = lgb.Dataset(x_val, y_val, reference=lgb_train)
    # 学習用パラメータ
    lgbm_params = {
        'objective': 'multiclass',
        'num_class': 4,
        }
    # 学習
    model = lgb.train(lgbm_params, lgb_train, valid_sets=lgb_eval)
    # 推定
    y_pred_proba_away = model.predict(x_test, num_iteration=model.best_iteration)
#     y_pred_away = np.argmax(y_pred_proba, axis=1)
    y_pred = get_y_pred2(y_pred_proba_home, y_pred_proba_away, y_test)
    # Accuracy の計算
    accuracy = sum(y_test == y_pred) / len(y_test)
    print('accuracy:', accuracy)

    sr.save([col_flg, pred_flg, accuracy])
```

・使用データ・・・①試合情報　＋　チーム順位  
・予測するもの・・・③2チームそれぞれで得点数の回帰⇨勝分敗

```python
def get_y_pred3(y_pred_home, y_pred_away):
    y_pred = []
    for i in range(len(y_pred_home)):
        if abs(y_pred_home[i]-y_pred_away[i])<0.5:
            y_pred += [0]
        elif y_pred_home[i]>y_pred_away[i]:
            y_pred += [1]
        else:
            y_pred += [2]
    return np.array(y_pred)
```

```python
def do_rightgbm_pred_flg3(col_flg):
    # 対象のデータを取得
    pred_flg = 3
    x_train, y_train, x_test, y_test = get_train_test(col_flg, pred_flg, False)
    x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, train_size=0.75)

    lgb_train = lgb.Dataset(x_train, y_train)
    lgb_eval = lgb.Dataset(x_val, y_val, reference=lgb_train) 
    # LightGBM parameters
    params = {
        'task': 'train',
        'boosting_type': 'gbdt',
        'objective': 'regression', # 回帰  
        'metric': {'rmse'}, # rsme(平均二乗誤差の平方根) 
    }
    # モデルの学習
    model = lgb.train(params,
                      train_set=lgb_train,
                      valid_sets=lgb_eval,
                      )
    # テストデータの予測
    y_pred_home = model.predict(x_test)
    
    # 対象のデータを取得
    x_train, y_train, x_test, y_test = get_train_test(col_flg, pred_flg, True)
    x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, train_size=0.75)

    lgb_train = lgb.Dataset(x_train, y_train)
    lgb_eval = lgb.Dataset(x_val, y_val, reference=lgb_train) 
    # LightGBM parameters
    params = {
        'task': 'train',
        'boosting_type': 'gbdt',
        'objective': 'regression', # 回帰  
        'metric': {'rmse'}, # rsme(平均二乗誤差の平方根) 
    }
    # モデルの学習
    model = lgb.train(params,
                      train_set=lgb_train,
                      valid_sets=lgb_eval,
                      )
    # テストデータの予測
    y_pred_away = model.predict(x_test)
    
    y_pred = get_y_pred3(y_pred_home, y_pred_away)
    # Accuracy の計算
    accuracy = sum(y_test == y_pred) / len(y_test)
    print('accuracy:', accuracy)

    sr.save([col_flg, pred_flg, accuracy])
```

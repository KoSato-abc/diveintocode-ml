次に、「自チーム」「相手チーム」「HomeAway」のカラムを  
「H_Team」(ホームチーム)「A_Team」(アウェイチーム)の2つにまとめます。

```python
H_teams = []
A_teams = []
for i, row in df2.iterrows():
    
    if row["HomeAway"] == "H":
        H_teams += [row["自チーム"]]
        A_teams += [row["相手"]]
    elif row["HomeAway"] == "A":
        H_teams += [row["相手"]]
        A_teams += [row["自チーム"]]
    else:
        print("例外")
df2 = df2.assign(H_Team = H_teams, A_Team = A_teams)
df2.drop(columns = ["自チーム","相手"], inplace = True)
```

次は「開催日」「年度」を「年月日」にまとめます。

```python
yyyymmdd_list = []
for i, row in df2.iterrows():
    
    year = str(row["年度"])
    mmdd = str(row["開催日"])
    
    if len(mmdd) == 3:
        month = mmdd[:1].zfill(2)
    else:
        month = mmdd[:2]
    day = mmdd[-2:]
    yyyymmdd_list += [year+month+day]

df2 = df2.assign(年月日 = yyyymmdd_list)
df2.drop(columns = ["開催日","年度"], inplace = True)
```

次に、「スコア」を  
「H_goal」「A_goal」「result」（得点、失点、試合結果）  
に分ていきます。  
「result」はtotoにならって、ホームチームの勝ち、負け、引き分けを1、2、0で表します。

```python
df2["スコア"].value_counts()
```

```python
H_goal_list = []
A_goal_list = []
result_list = []
for i, row in df2.iterrows():
    left = int(row["スコア"].split('-')[0])
    right = int(row["スコア"].split('-')[1])
    if row["HomeAway"] == "H":
        H_goal = left
        A_goal = right
    else:
        H_goal = right
        A_goal = left
    if H_goal < A_goal:
        result = "2"
    elif H_goal > A_goal:
        result = "1"
    else:
        result = "0"
    
    H_goal_list += [H_goal]
    A_goal_list += [A_goal]
    result_list += [result]

df2 = df2.assign(H_goal = H_goal_list, A_goal = A_goal_list, H_result = result_list)
df2.drop(columns = ["スコア", "HomeAway"], inplace = True)
```

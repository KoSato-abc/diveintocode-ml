## スクレイピングして取得した各csvファイルを一つのcsvファイルにまとめる

```python
import pandas as pd
import numpy as np
```

```python
df_stats = pd.read_csv('02_01_footboll-lab/csv/data_by_football-lab.csv', index_col=0)
df_schedule = pd.read_csv('02_02_j-league-data-site/csv/j_match_schedule.csv', index_col=0)
df_rank = pd.read_csv('02_02_j-league-data-site/csv/j_rank_table.csv', index_col=0)
df_profiles = pd.read_csv('02_98_others/J_PlayerProfiles_from_official.csv', index_col=0)
df_starting_member = pd.read_csv('02_98_others/StartingMember_sum.csv', index_col=0)
```

最終的には一つの行ごとに、年月日、ホームチーム情報、アウェイチーム情報を持ちたいと思います。  
それぞれのデータのチーム名と日付をキーにして結合できそうです。  
しかし、チーム名の表記がデータごとにバラバラなので、まずはそこを合わしていきます。

```python
team_names = pd.read_csv('TEAM_NAMES.csv').values.tolist()
```

```python
team_names
```

```python
def get_rename_team(df, tar_col):
    df = df.copy()
    res_val_list = []

    for i, row in df.iterrows():
        count_change = 0
        for i, team_list in enumerate(team_names):
            if row[tar_col] in team_list:
                res_val_list += [team_list[0]]
                count_change += 1
        if count_change == 1:
            pass
        else:
            print(f"例外あり：{row[tar_col]}")
    
    return res_val_list
```

```python
new_name_list = get_rename_team(df_stats, "H_Team")
df_stats.drop(columns = ["H_Team"], inplace = True)
df_stats = df_stats.assign(H_Team = new_name_list)
df_stats
```

```python
new_name_list = get_rename_team(df_stats, "A_Team")
df_stats.drop(columns = ["A_Team"], inplace = True)
df_stats = df_stats.assign(A_Team = new_name_list)
df_stats
```

```python
new_name_list = get_rename_team(df_schedule, "ホーム")
df_schedule.drop(columns = ["ホーム"], inplace = True)
df_schedule = df_schedule.assign(ホーム = new_name_list)
df_schedule
```

```python
new_name_list = get_rename_team(df_schedule, "アウェイ")
df_schedule.drop(columns = ["アウェイ"], inplace = True)
df_schedule = df_schedule.assign(アウェイ = new_name_list)
df_schedule
```

```python
new_name_list = get_rename_team(df_rank, "チーム")
df_rank.drop(columns = ["チーム"], inplace = True)
df_rank = df_rank.assign(チーム = new_name_list)
df_rank
```

```python
new_name_list = get_rename_team(df_profiles, "最終所属")
df_profiles.drop(columns = ["最終所属"], inplace = True)
df_profiles = df_profiles.assign(最終所属 = new_name_list)
df_profiles
```

```python
new_name_list = get_rename_team(df_starting_member, "チーム名")
df_starting_member.drop(columns = ["チーム名"], inplace = True)
df_starting_member = df_starting_member.assign(チーム = new_name_list)
df_starting_member
```

それぞれのデータを結合していきます。

```python
import numpy as np

df = df_stats.copy()
df_rank_columns = df_rank.columns.values

year_list = []
for i, row in df.iterrows():
    year = str(row["年月日"])[:4]
    year_list += [int(year)]
df = df.assign(year = year_list)

df_rank_columns_H = df_rank_columns.copy()
for i, col in enumerate(df_rank_columns):
    df_rank_columns_H[i] =  "H_" + col
df_rank_H = df_rank.set_axis(df_rank_columns_H, axis='columns')

df_rank_columns_A = df_rank_columns.copy()
for i, col in enumerate(df_rank_columns):
    df_rank_columns_A[i] =  "A_" + col
df_rank_A = df_rank.set_axis(df_rank_columns_A, axis='columns')

df_marge = pd.merge(df, df_rank_H, 
         left_on=['year', '節', 'H_Team'],
         right_on=['H_year', 'H_節', 'H_チーム']).drop(columns=['H_year', 'H_節', 'H_チーム'])

df_marge = pd.merge(df_marge, df_rank_A, 
         left_on=['year', '節', 'A_Team'],
         right_on=['A_year', 'A_節', 'A_チーム']).drop(columns=['A_year', 'A_節', 'A_チーム'])

df_marge.rename(columns={'H_カテゴリ': 'カテゴリ'}, inplace=True)
df_marge.drop(columns=['A_カテゴリ'], inplace=True)
```

```python
df_marge
```

```python
df = df_starting_member.copy()

key_list = []
player_list = []
coach_list = []
referee_list = []
table = str.maketrans({
    v: '' for v in '\u3000 \x0c\x0b\t'
})
for i, row in df.iterrows():
    date = str(row["日付"])
    team = row["チーム"]
    
    key_list += [date+team]
    player_list += [row["選手"].translate(table)]
    coach_list += [row["監督"].translate(table)]
    referee_list += [row["主審"].translate(table)]
    
df.drop(columns = ["選手", "監督", "主審"], inplace = True)
df = df.assign(key = key_list, 選手 = player_list, 監督 = coach_list, 主審 = referee_list)
```

```python
df.info()
```

```python
key_list = df["key"].value_counts().index
columns = ["日付", "カテゴリ", "節", "チーム", "監督", "気温", "湿度", "主審"
           , "ポジション1", "選手1", "ポジション2", "選手2", "ポジション3", "選手3", "ポジション4", "選手4", "ポジション5", "選手5"
          , "ポジション6", "選手6", "ポジション7", "選手7", "ポジション8", "選手8", "ポジション9", "選手9", "ポジション10", "選手10"
          , "ポジション11", "選手11"]
new_df_starting_member = None
for _, key in enumerate(key_list):
    df_tmp = df[df["key"]==key]
    if df_tmp.shape[0] != 11:
        print("NG : ", key)
    player_list = []
    position_list = []
    position_list += df_tmp[df_tmp["ポジション"]=="GK"]["ポジション"].tolist()
    position_list += df_tmp[df_tmp["ポジション"]=="DF"]["ポジション"].tolist()
    position_list += df_tmp[df_tmp["ポジション"]=="MF"]["ポジション"].tolist()
    position_list += df_tmp[df_tmp["ポジション"]=="FW"]["ポジション"].tolist()
    player_list += df_tmp[df_tmp["ポジション"]=="GK"]["選手"].tolist()
    player_list += df_tmp[df_tmp["ポジション"]=="DF"]["選手"].tolist()
    player_list += df_tmp[df_tmp["ポジション"]=="MF"]["選手"].tolist()
    player_list += df_tmp[df_tmp["ポジション"]=="FW"]["選手"].tolist()
    pp_list = []
    for i in range(11):
        pp_list += [position_list[i]] + [player_list[i]]
    
    date = df_tmp["日付"].values[0]
    category = df_tmp["カテゴリ"].values[0]
    section = df_tmp["節"].values[0]
    team = df_tmp["チーム"].values[0]
    coach = df_tmp["監督"].values[0]
    temperature = df_tmp["気温"].values[0]
    humidity = df_tmp["湿度"].values[0]
    referee = df_tmp["主審"].values[0]
    
    new_row = [date, category, section, team, coach, temperature, humidity, referee] + pp_list
    new_df = pd.DataFrame([new_row], columns=columns)
    new_df_starting_member = pd.concat([new_df_starting_member, new_df])
```

```python
columns = new_df_starting_member.columns.values
df_sm = new_df_starting_member.copy()
df_sm["節"] = df_sm["節"].astype(int)
df_marge2 = df_marge.copy()

columns_H = columns.copy()
for i, col in enumerate(columns):
    columns_H[i] =  "H_" + col
df_sm_H = df_sm.set_axis(columns_H, axis='columns')

columns_A = columns.copy()
for i, col in enumerate(columns):
    columns_A[i] =  "A_" + col
df_sm_A = df_sm.set_axis(columns_A, axis='columns')

df_marge2 = pd.merge(df_marge2, df_sm_H, 
         left_on=['年月日', '節', 'H_Team'],
         right_on=['H_日付', 'H_節', 'H_チーム'], how='left').drop(columns=['H_日付', 'H_節', 'H_チーム'])

df_marge2 = pd.merge(df_marge2, df_sm_A, 
         left_on=['年月日', '節', 'A_Team'],
         right_on=['A_日付', 'A_節', 'A_チーム'], how='left').drop(columns=['A_日付', 'A_節', 'A_チーム'])

df_marge2.rename(columns={'H_気温': '気温','H_湿度':'湿度','H_主審':'主審'}, inplace=True)
df_marge2.drop(columns=['A_気温', 'A_湿度', 'A_主審', 'H_カテゴリ', 'A_カテゴリ', 'year'], inplace=True)
```

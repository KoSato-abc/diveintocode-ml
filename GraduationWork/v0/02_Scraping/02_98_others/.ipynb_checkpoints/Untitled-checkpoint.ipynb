```python
import re

category_list = []
section_list = []
for i, row in df.iterrows():
    tar_str = row["試合情報"]
    # 節
    section_m = re.findall('第(.*)節', tar_str)
    section_list += section_m 
    # カテゴリ
    category_m = re.findall('Ｊ(\d)', tar_str)
    if len(category_m) == 0:
        category_m = re.findall('ディビジョン(\d)', tar_str)
    category = "J" + category_m[0]
    category_list += [category] 
    
df = df.assign(節 = section_list, カテゴリ = category_list)
df.drop(columns = ["試合情報"], inplace = True)
```

```python
df
```

```python
tar_col_list = ["日付", "湿度"]
dic = {'/':None, '%':None}
for _, col in enumerate(tar_col_list):
    df[col] = df[col].apply(lambda x : x.translate(str.maketrans(dic)))
```

次に、「match_schedule_original.csv」について作業します。

```python
df = df_schedule.copy()
df.info()
```

```python
df
```

欠損値を確認します。

```python
# 欠損値が含まれている列と数を抽出
drop_col = []
s = df.isnull().sum()
for i, n in enumerate(s.values):
    if(0<n):
        drop_col += [s.index[i]]
        print(s.index[i], n)
```

「節」に欠損値があるのは想定外なので確認してみます。

```python
df[df["節"].isna()]["大会"].value_counts()
```

どうやら、Jリーグ以外の大会での試合の場合に欠損しているだけのようです。  
これは問題ないので、「節」以外の欠損値が含まれる列を削除します。

```python
df2 = df.drop(columns = ["Unnamed: 0", "K/O時刻", "入場者数","インターネット中継・TV放送", "Team"])
```

```python
df2
```

次に、「節」の "第○節第×日" を "○" にします。

```python
import re

section_list = []
for i, row in df2.iterrows():
    tar_str = row["節"]
    if tar_str is np.nan:
        section_list += [0]
    else:
        section_m = re.findall('第(.*)節', tar_str)
        if len(section_m) == 0:
            section_list += [0]
        else:
            section_list += section_m

df2.drop(columns = ["節"], inplace = True)
df2 = df2.assign(節 = section_list)
```

```python
df2
```

次に、年度、試合日を yyyyMMdd の形式にします。

```python
import re

MMdd_list = []
for i, row in df2.iterrows():
    year = str(row["年度"])
    tar_str = row["試合日"]
    if tar_str == "未定":
        MMdd_list += [tar_str]
    else:
        month = re.findall('(.*)/', tar_str)[0]
        day = re.findall('/(.*)\(', tar_str)[0]
        MMdd_list += [year+month+day]

df2.drop(columns = ["試合日", "年度"], inplace = True)
df2 = df2.assign(試合日 = MMdd_list)
```

```python
# 重複の削除
df2.drop_duplicates(inplace=True)
```

```python
df2
```

```python
columns_list = ['試合日', '節', '大会', 'ホーム', 'スコア', 'アウェイ', 'スタジアム']
df2 = df2.reindex(columns=columns_list)
df2
```

ここら辺で終わりにします。

```python
df2.to_csv("j_match_schedule.csv")
```

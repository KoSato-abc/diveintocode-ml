```python
def _df_rename_team(df, tar_col):
    df = df.reset_index(drop = True).copy()
    new_name_list = []
    M_TEAM_NAMES = pd.read_csv('data/other/TEAM_NAMES.csv').values.tolist()
    
    for i, row in df.iterrows():
        count_change = 0
        
        if row[tar_col] is np.nan:
            new_name_list += ['-']
            count_change += 1
        else:
            for i, team_list in enumerate(M_TEAM_NAMES):
                if row[tar_col] in team_list:
                    new_name_list += [team_list[0]]
                    count_change += 1

        if count_change == 1:
            pass
        elif count_change == 0:
            new_name_list += [row[tar_col]]
            print(f"TEAM_NAMES.csvに記載なし：{row[tar_col]}")
        else:
            print(f"TEAM_NAMES.csvを修正してください：{row[tar_col]}")
            
    df.drop(columns = [tar_col], inplace = True)

    df_tmp = pd.DataFrame(new_name_list, columns = [tar_col])
        
    df = pd.concat([df, df_tmp],axis=1)
    return df
```

```python
def get_last_team(player, df_marge):
    columns1 = ['年月日', 'H_Team', 'A_Team']
    columns2 = ['H_選手1', 'H_選手2', 'H_選手3', 'H_選手4', 'H_選手5', 'H_選手6', 'H_選手7', 'H_選手8', 'H_選手9', 'H_選手10', 'H_選手11']
    columns3 = ['A_選手1', 'A_選手2', 'A_選手3', 'A_選手4', 'A_選手5', 'A_選手6', 'A_選手7', 'A_選手8', 'A_選手9', 'A_選手10', 'A_選手11']
    df = df_marge.copy()
    
    df = df[columns1 + columns2 + columns3]
    last_play_match = df[(df == player).sum(axis='columns').astype(bool)].sort_values('年月日', ascending=False)[:1]
    
#     return last_play_match
    
    if (last_play_match[columns2]==player).sum(axis='columns').astype(bool).values[0]:
        last_team = last_play_match['H_Team'].values[0]
    else:
        last_team = last_play_match['A_Team'].values[0]
    return last_team
```

**【問題1】乱数の作成**

```
import numpy as np
np.random.seed(0)

mu = [-3, 0]
sigma = [[1.0, 0.8], [0.8, 1.0]]
 
# 2次元正規乱数を500個生成
values = np.random.multivariate_normal(mu, sigma, 500)
```

**【問題2】散布図による可視化**

```
import matplotlib.pyplot as plt

x = [r[0] for r in values]
y = [r[1] for r in values]

# 散布図を描画
plt.scatter(x,y)
```

**【問題3】ヒストグラムによる可視化**

```
plt.hist(x, bins=100)
plt.xlim(-5, 2)
```

```
plt.hist(y, bins=100)
plt.xlim(-5, 2)
```

**【問題4】データの追加**

```
mu = [0, -3]
 
# 2次元正規乱数を500個生成
values2 = np.random.multivariate_normal(mu, sigma, 500)

x2 = [r[0] for r in values2]
y2 = [r[1] for r in values2]

# 散布図を描画
plt.scatter(x,y,label='1')
plt.scatter(x2,y2,label='2')
plt.legend()
```

**【問題5】データの結合**

```
values_vstack = np.vstack([values,values2])
print(values_vstack.shape)
```

**【問題6】ラベル付け**

```
values_label = np.insert(values, 2, 0, axis=1)
values2_label = np.insert(values2, 2, 1, axis=1)

values_vstack = np.vstack([values_label,values2_label])
print(values_vstack.shape)
```
